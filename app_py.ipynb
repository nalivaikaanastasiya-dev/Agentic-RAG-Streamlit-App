```python
# app.py
import streamlit as st
from langchain.document_loaders import PyPDFLoader
from agent_rag_pipeline import AgenticRAG
from dotenv import load_dotenv

# -----------------------------
# Load environment keys
# -----------------------------
load_dotenv()

st.set_page_config(page_title="Agentic RAG", layout="wide")
st.title("Agentic RAG Chat")

# Upload PDF documents
uploaded_file = st.file_uploader("Upload PDF", type="pdf")
if uploaded_file:
    loader = PyPDFLoader(uploaded_file)
    docs = loader.load()
    st.success(f"Loaded {len(docs)} documents")

    # Initialize agent
    agent = AgenticRAG(docs)

    # User query
    query = st.text_input("Ask a question:")
    if query:
        answer = agent.get_answer(query)
        st.write("Answer:")
        st.write(answer)

# Optional: show notebook preview
try:
    with open("agentic_rag.ipynb", "r", encoding="utf-8") as f:
        st.subheader("Agent Notebook Preview")
        st.text(f.read()[:1000])
except Exception as e:
    st.warning(f"Could not load agentic_rag.ipynb: {e}")
```
