```python
import streamlit as st
from agent_rag_pipeline import AgenticRAG
from langchain.document_loaders import TextLoader
from dotenv import load_dotenv

load_dotenv()

st.set_page_config(page_title="Agentic RAG", layout="wide")
st.title("Agentic RAG Chat")

# Load example documents
loader = TextLoader("example_docs.txt")
docs = loader.load()

# Initialize the Agentic RAG agent
agent = AgenticRAG(docs)

# Streamlit input
query = st.text_input("Ask a question:")
if query:
    answer = agent.get_answer(query)
    st.write("Answer:")
    st.write(answer)

# Preview agent notebook
try:
    with open("agentic_rag.ipynb", "r", encoding="utf-8") as f:
        st.subheader("Agent Notebook Preview")
        st.text(f.read()[:1000])
except Exception as e:
    st.warning(f"Could not load agentic_rag.ipynb: {e}")
```
